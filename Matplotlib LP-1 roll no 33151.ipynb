```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
%matplotlib inline
```

# Sales Data Visualization

```python
Sales = pd.read_csv("company_sales_data.csv")
Sales.head()
```

```python
#Q1

plt.plot(Sales['month_number'],Sales['total_profit'])
plt.title("Company profit per month")
plt.xlabel("Month Number")
plt.ylabel("Total profit")
```

```python
#Q2

plt.plot(Sales['month_number'],Sales['total_units'], color='red', marker='o', linestyle='dashed',linewidth=3, markerfacecolor='black', markersize=10,label='Profit data of last year')
plt.title("Comapny sales data of last year")
plt.legend(loc="lower right")
plt.xlabel("Month Number")
plt.ylabel("Total units")
```

```python
#Q3

plt.scatter(Sales['month_number'],Sales['toothpaste'])
plt.grid(linestyle='-',linewidth=1)
plt.xticks(np.arange(0, 13, 1))
plt.title("Tooth paste Sales data")
plt.xlabel("Month Number")
plt.ylabel("Number of units sold")
```

```python
#Q4

Sales.plot(x='month_number', y=['facecream','facewash'], kind="bar",label = ['Face Cream sales data','Face Wash sales data'])
plt.grid(linestyle='-',linewidth=1)
plt.title("Fash wash and Face cream Sales data")
plt.xlabel("Month Number")
plt.ylabel("Number of units sold")
plt.legend(loc="upper left")
```

```python
#Q5

plt.bar(Sales['month_number'],Sales['bathingsoap'])
plt.grid(linestyle='-',linewidth=1)
plt.xticks(np.arange(0, 13, 1))
plt.title("Bathing soap Sales data")
plt.xlabel("Month Number")
plt.ylabel("Number of units sold")
```

```python
#Q6

profit_List = Sales ['total_profit'].tolist()
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(profit_List, profit_range, label = 'Profit data')
plt.xticks(profit_range)
plt.legend(loc='upper left')
plt.title("Profit Data")
plt.xlabel("profit range in dollar")
plt.ylabel("actual profit in dollar")
```

```python
#Q7

Months  = Sales['month_number'].tolist()

face_Cream_Data   = Sales ['facecream'].tolist()
face_Wash_Data   = Sales ['facewash'].tolist()
tooth_Paste_Data = Sales ['toothpaste'].tolist()
bathing_soap_Data   = Sales ['bathingsoap'].tolist()
shampoo_Data   = Sales ['shampoo'].tolist()
moisturizer_Data = Sales ['moisturizer'].tolist()

plt.plot([],[],color='purple', label='Face Cream', linewidth=5)
plt.plot([],[],color='blue', label='Face Wash', linewidth=5)
plt.plot([],[],color='red', label='Tooth paste', linewidth=5)
plt.plot([],[],color='black', label='Bathing soap', linewidth=5)
plt.plot([],[],color='green', label='Shampoo', linewidth=5)
plt.plot([],[],color='yellow', label='Moisturizer', linewidth=5)


plt.stackplot(Months, face_Cream_Data, face_Wash_Data, tooth_Paste_Data, bathing_soap_Data, shampoo_Data, moisturizer_Data, colors=['purple','cyan','red','black','green','yellow'])

plt.title("All product sales data using stack plot")
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
```

# Tempeature Dataset Visualization

```python
Temp = pd.read_csv("temperatures.csv")
Temp.head()
```

```python
# Graph No 1
plt.plot(Temp['YEAR'],Temp['ANNUAL'],color='red')
plt.title("Average Temperature Per Year")
plt.xlabel("YEAR")
plt.ylabel("Temperature")
```

Above Plot is a Graph which reprsents the change in Average temp per year from 1901 to 2016

```python
# Graph No 2
plt.plot(Temp['YEAR'],Temp['JAN'],label='JAN')
plt.plot(Temp['YEAR'],Temp['FEB'],label="FEB")
plt.plot(Temp['YEAR'],Temp['MAR'],label='MAR')
plt.plot(Temp['YEAR'],Temp['APR'],label='APR')
plt.plot(Temp['YEAR'],Temp['MAY'],label='MAY')
plt.plot(Temp['YEAR'],Temp['JUN'],label='JUN')
plt.plot(Temp['YEAR'],Temp['JUL'],label='JUL')
plt.plot(Temp['YEAR'],Temp['AUG'],label='AUG')
plt.plot(Temp['YEAR'],Temp['SEP'],label='SEP')
plt.plot(Temp['YEAR'],Temp['OCT'],label='OCT')
plt.plot(Temp['YEAR'],Temp['NOV'],label='NOV')
plt.plot(Temp['YEAR'],Temp['DEC'],label='DEC')
plt.title("Temperature for each year each month")
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.legend(loc='upper left')
```

Above grpah has the line plot of all the months temperature from year 1901 to 2016

```python
Temporary1=Temp[Temp.YEAR==2016]
Temporary1
```

```python
# Graph No 3
Temporary1.plot(x='YEAR',y=['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC'],kind = 'bar')
plt.title('Temperature for each Month in 2016')
plt.ylabel('Temperature')
```

Above Bar plot shows the temperature in each month of 2016(which the latest year in dataset)

As from the above plot we can clearly say that may was the hottest month  and Jan was the coldest month in 2016

```python
# Graph No 4
plt.plot(Temp['YEAR'],Temp['JAN-FEB'],label='JAN-FEB')
plt.plot(Temp['YEAR'],Temp['MAR-MAY'],label='MAR-MAY')
plt.plot(Temp['YEAR'],Temp['JUN-SEP'],label='JUN-SEP')
plt.plot(Temp['YEAR'],Temp['OCT-DEC'],label='OCT-DEC')
plt.title("Temperature for each Quater of Year")
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.legend(loc='upper left')
```

Above graph shows the Quarterly temp Average per year

```python
# Graph Nop 5
plt.plot(Temp['YEAR'],Temp['JAN'],color='red')
plt.title("Average Temperature Per Year For January")
plt.xlabel("YEAR")
plt.ylabel("Temperature")
```

```python
#Graph no 6
plt.plot(Temp['YEAR'],Temp['MAY'],color='red')
plt.title("Average Temperature Per Year For May")
plt.xlabel("YEAR")
plt.ylabel("Temperature")
```

In Graph no 5 and 6 we show the change in temperature of two months January and May which are the coldest and hottest months in a year respectively
